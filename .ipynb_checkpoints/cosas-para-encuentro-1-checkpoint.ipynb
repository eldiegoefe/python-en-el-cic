## Listas ordenadas

Cada item se coloca con un prefijo numérico y un punto. Pueden ser todos los números iguales: 0.

La indentación determina items de la misma categoría

1. Item 1
1. Item 2
33. Item 3
    0. Subitem 1
    0. Subitem 2
       0. Subsubitem 1
           0. Subsubsubsub
    0. Subitem 3

### Listas desordenadas (y combinadas)

Cada item se coloca con un prefijo: - * +

La indentación determina items de la misma categoría

- Item 1
1. Item 2
33. Item 3
    - Subitem 1
    0. Subitem 2
       0. Subsubitem 1
       0. Subsubitem 2
       0. Subsubitem 3
           0. Subsubsubsub
               * item
               * item
               * item
    0. Subitem 3

# Minicurso de LaTeX (1/2)

- **Ecuaciones**. Se escriben entre \$ o entre \$\$:
    - \$ ecuación "inline" \$
    - \$\$ ecuación separada \$\$
--- 

- Letras griegas:
    
| Código | Apariencia |
|:-------: | :--------: |
| \alpha   | $\alpha$ |
| \beta    | $\beta$  |
| \pi      | $\pi$    |


- Indices:
    - Superíndices: ^
    - Subíndice: _
    
| Código     | Apariencia |
|:-------:   | :--------: |
| I_calcio^2   | $I_calcio^2$ |
| I_{calcio}^2 | $I_{calcio}^2$  |


- fracciones:

| Código | Apariencia |
| :-------: | :--------: |
| \frac{num}{den} | $\frac{num}{den}$ |


# Minicurso de LaTeX (2/2)

- paréntesis, llaves, corchetes:
    - \left( y \right) (se escalan de acuerdo al contenido)
        - \left( \right): $\left( \right)$
        - \left[ \right]: $\left[ \right]$
        - \left\\{ \right\\}: $\left\{ \right\}$   (atención al escape: \\)
        - ejemplo de escalado: $\left( \int^{\infty}_{n = -\infty} \frac{1}{n^2}\right)$

- funciones estándar:
    - seno: \sin $\rightarrow \sin \left( 2 \pi \right)$
    - logaritmo: \ln $\rightarrow \ln$

- flechas:
    - \rightarrow $\rightarrow$
    - \leftarrow $\leftarrow$

### Potencial de equilibrio de un ion (ecuación de Nernst)

E_K+ = RT / zF * ln ([K+]o / [K+]i)

$$E_{K^+} = \frac{RT}{zF} \ln \frac{[K^+]_{o}}{[K^+]_{i}}$$
