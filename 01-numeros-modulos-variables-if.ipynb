# Formato de salida en la orden Print

- una referencia: [String Formatting Cookbook](https://mkaz.github.io/2012/10/10/python-string-format)

````python
bidon = 15        # en litros 
termo = 0.9

bidon = bidon - 2 * termo
print("En el bidon y en el termo quedan respectivamente: ", bidon, termo)
print("En el bidón quedan {1} litros y en el termo hay {0} litros".format(termo, bidon))
print("En el bidón quedan {1:.3f} litros y en el termo hay {0:3.1e} litros.".format(termo, bidon))
````

## if...else...

``` python
if condicion:
    orden 1
    orden 2
    ...
else:
    orden a
    orden b
    orden c
    ...
```

- **los bloques se identifican por la indentación**

## If evalúa condiciones lógicas
### la condición siempre debe evaluar a True o False

+ puede ser una expresión: 
    
    ``` python 
    if valor > umbral:
        bloque de ordenes
    ```
    
+ pueden ser varias expresiones: 
    
    ``` python 
    if (valor > piso) and (valor < techo):
        bloque de ordenes
    ```
    
+ puede ser una variable booleana: 
    
    ``` python
    superaUmbral = valor > umbral
    if superaUmbral:
        bloque de ordenes
     ```

````python
bidon = 15
termo = 0.9

if termo < bidon:
    print("El termo es mas pequeño que el bidón")
    print("¡Es obvio!")
else:
    print("¡Tenés un termo grande!")
````

# Anidando condiciones (elif)

````python
bidon = 15
termo = 0.9

if termo < bidon:
    print("El termo es mas pequeño que el bidón")
    print("¡Es obvio!")
elif (termo - bidon) < 0.05:
    print("Tenés un termo raro: ¡es igual al bidón!")
    print("¡muy raro!")
else:
    print("El termo es más grande que el bidón")

print("\nhasta la vista, baby!")
````
