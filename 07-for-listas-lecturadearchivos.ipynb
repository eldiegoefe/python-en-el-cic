# Bucles

- Ejecución del mismo conjunto de órdenes un número prefijado de veces
- Sintaxis (hay variantes):

``` python

for x in range(num):
     orden 1
     orden 2
     ...
     orden n
```

# Sumar los enteros del 1 al 100

``` python
# inicializamos dos variables
acumulador = 0
n = 1

# repetimos hasta que n = 100
acumulador = acumulador + n  
n = n + 1    # n = 2
acumulador = acumulador + n  
n = n + 1    # n = 3
acumulador = acumulador + n
n = n + 1    # n = 4
```

*Tedio total ----> ¡chau programación!*

## Con la instrucción FOR las iteraciones se condensan:

````python
acumulador = 0
n_max = 100

for n in range(n_max):                 # range en python 3 es un tipo de datos en sí mismo
    # print("---------------------> n vale: ", n)
    acumulador = acumulador + n

print("La suma de los primeros {0} enteros da {1}".format(n, acumulador))
````

# Lectura de un registro

- Lectura de los datos desde un archivo
- Conversión a dos listas numéricas (una para cada columna)
- Primera columna: tiempos
- Segunda columna: fluorescencia
- Cada linea tiene caracteres raros ('\n', '\t', etc)
- Los datos se deben convertir de str a float

# Variable con el path y el nombre del archivo a leer

````python
miDirectorio = './datos/'
nombreArchivo = 'registro1.txt'
archivoDatos = miDirectorio + nombreArchivo
archivoDatos
````

# Apertura, lectura y cierre del archivo

- se debe abrir y cerrar el archivo
- la función "readlines" lee el archivo
- convierte cada linea en un elemento de una lista

````python
file = open(archivoDatos, 'r')
# print("----------------> Tipo de 'file': ", type(file))
lineas = file.readlines()
file.close()
````

# ¿Qué había adentro del archivo?

````python
print(lineas[0])
print(repr(lineas[0]))
type(lineas[0])
````

````python
for t in lineas:
    listaPequenia = t.split('\t')
    valorIzquierdo = listaPequenia[0]
    valorConPunto = valorIzquierdo.replace(',', '.')
    print(repr(valorConPunto))
    print()
````

````python
for t in lineas:
    listaPequenia = t.split('\t')
    valorIzquierdo = listaPequenia[0]
    valorConPunto = valorIzquierdo.replace(',', '.')
    numeroFinal = float(valorConPunto)
    print(repr(numeroFinal))
    print()
````

````python
for t in lineas:
    numeroFinal=float(t.split('\t')[0].replace(',', '.'))
    print(repr(numeroFinal))
    print()
````

````python
listaTiempo = []

for t in lineas:
    numeroFinal=float(t.split('\t')[0].replace(',', '.'))
    listaTiempo.append(numeroFinal)

print(listaTiempo)
````

# List Comprehension

``` python

identificador = [f(x) for x in lista]
```

````python
listaIdentidad = [x for x in range(20)]
print(listaIdentidad)
````

````python
cuadrados = [t**2 for t in range(20)]
print(cuadrados)
````

````python
nuevaListaTiempo = [float(t.split('\t')[0].replace(',', '.')) for t in lineas]
print(nuevaListaTiempo)
````
